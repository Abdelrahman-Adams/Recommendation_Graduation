```python
import pandas as pd

df = pd.read_csv('/content/job_postings.csv')
filtered_data = df[['title', 'description']]
filtered_data
```

```python
filtered_data['description'].values.tolist()
```

```python
from keybert import KeyBERT
keywords=[]
model = KeyBERT(model='BERT-large-uncased')
for doc in filtered_data['description']:

  keyword = model.extract_keywords(doc, top_n=5, keyphrase_ngram_range=(1, 2), stop_words='english')
  keywords.append(keyword)
for keyword, score in keywords:
    print(f"Keyword: {keyword}")
    print(f"Score: {score}")
```

```python
# Initialize the KeyBERT model
kw_model = KeyBERT('distilbert-base-nli-mean-tokens')

# Function to extract keywords from the description column
def extract_skills(row):
    description = row['description']
    keywords = kw_model.extract_keywords(description, keyphrase_ngram_range=(1, 2), stop_words='english')
    return [kw for kw, _ in keywords]

# Apply the function to create a new column "skills"
filtered_data['description'] = filtered_data.apply(extract_skills, axis=1)

# Display the DataFrame with the added "skills" column
filtered_data
```

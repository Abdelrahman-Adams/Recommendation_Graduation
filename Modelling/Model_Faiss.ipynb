```python
print(updated_data['combined_values'].isnull().sum())
updated_data = updated_data.reset_index(drop=True)
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-mpnet-base-v2')

# #Our sentences we like to encode
# sentences = ['This framework generates embeddings for each input sentence',
#     'Sentences are passed as a list of string.',
#     'The quick brown fox jumps over the lazy dog.']

#Sentences are encoded by calling model.encode()
embeddings = model.encode(updated_data['combined_values'])
#Print the embeddings
```

```python
for sentence, embedding in zip(updated_data['combined_values'], embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
    break
```

```python
import pandas as pd
import numpy as np
import faiss

# Initialize the index in Faiss
d = embeddings.shape[1]  # dimension
index = faiss.IndexFlatL2(d)
index.add(embeddings)

faiss.write_index(index, 'faiss_index.index')

model.save('sentence_transformer_model')

# Query vector (for demonstration purposes, you may replace it with your actual query vector)
query_vector = model.encode(updated_data['combined_values'][0])

# Perform a k-nearest neighbors search
k = 5
distances, indices = index.search(np.expand_dims(query_vector, axis=0), k)

# Display the results
# print("Query Vector:")
# print(query_vector)
print("\nNearest Neighbors:")
for i, (dist, ind) in enumerate(zip(distances[0], indices[0])):
    print(f"Rank {i+1}: Index {ind}, Distance {dist}")
    print(updated_data.loc[ind])  # Display the corresponding DataFrame record
    print("")
```

```python
import gdown
import pandas as pd

# Google Drive file ID
file_id = '14bCbOcmBdpF9k36jyoE5jRDe-wZe5UiS'

# Define the file path where you want to save the downloaded file
file_path = '/content/dataset.csv'
# Download the file from Google Drive using its file ID
gdown.download(f'https://drive.google.com/uc?id={file_id}', file_path, quiet=False)

# Read the downloaded data into a Pandas DataFrame
combined_data = pd.read_csv(file_path).astype(str)

```

```python
combined_data
```

```python
# Load the SentenceTransformer model from a file
loaded_model = SentenceTransformer('/content/drive/MyDrive/Graduation_Data/ModelAndWeights')

# Load the Faiss index from a file
loaded_index = faiss.read_index('/content/drive/MyDrive/Graduation_Data/ModelAndWeights/faiss_index.index')
```

```python
# Query vector (for demonstration purposes, you may replace it with your actual query vector)
query_vector = loaded_model.encode(combined_data['combined_values'][15])

# Perform a k-nearest neighbors search
k = 5
distances, indices = loaded_index.search(np.expand_dims(query_vector, axis=0), k)

# Display the results
# print("Query Vector:")
# print(query_vector)
print("\nNearest Neighbors:")
for i, (dist, ind) in enumerate(zip(distances[0], indices[0])):
    print(f"Rank {i+1}: Index {ind}, Distance {dist}")
    print(combined_data.loc[ind])  # Display the corresponding DataFrame record
    print("")
```

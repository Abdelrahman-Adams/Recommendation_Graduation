```python
import pandas as pd

data=pd.read_csv('/content/FinalTable.csv').astype(str)
data
```

```python
# data['Application Score'] = data['Application Score'].astype(str)
# data=data.drop(['js_id'],axis=1)
data['combined_values'] = data.apply(lambda row: ' </s> '.join(row), axis=1)

# Display the updated DataFrame
data['combined_values'][0]
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-mpnet-base-v2')

# #Our sentences we like to encode
# sentences = ['This framework generates embeddings for each input sentence',
#     'Sentences are passed as a list of string.',
#     'The quick brown fox jumps over the lazy dog.']

#Sentences are encoded by calling model.encode()
embeddings = model.encode(data['combined_values'])
#Print the embeddings
```

```python
for sentence, embedding in zip(data['combined_values'], embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
    break
```

```python
import pandas as pd
import numpy as np
import faiss

# Initialize the index in Faiss
d = embeddings.shape[1]  # dimension
index = faiss.IndexFlatL2(d)
index.add(embeddings)

faiss.write_index(index, 'faiss_index.index')

model.save('sentence_transformer_model')

# Query vector (for demonstration purposes, you may replace it with your actual query vector)
query_vector = model.encode(data['combined_values'][0])

# Perform a k-nearest neighbors search
k = 5
distances, indices = index.search(np.expand_dims(query_vector, axis=0), k)

# Display the results
# print("Query Vector:")
# print(query_vector)
print("\nNearest Neighbors:")
for i, (dist, ind) in enumerate(zip(distances[0], indices[0])):
    print(f"Rank {i+1}: Index {ind}, Distance {dist}")
    print(data.loc[ind])  # Display the corresponding DataFrame record
    print("")
```

```python
# Load the SentenceTransformer model from a file
loaded_model = SentenceTransformer('sentence_transformer_model')

# Load the Faiss index from a file
loaded_index = faiss.read_index('faiss_index.index')
# Query vector (for demonstration purposes, you may replace it with your actual query vector)
query_vector = loaded_model.encode(data['combined_values'][0])

# Perform a k-nearest neighbors search
k = 5
distances, indices = loaded_index.search(np.expand_dims(query_vector, axis=0), k)

# Display the results
# print("Query Vector:")
# print(query_vector)
print("\nNearest Neighbors:")
for i, (dist, ind) in enumerate(zip(distances[0], indices[0])):
    print(f"Rank {i+1}: Index {ind}, Distance {dist}")
    print(data.loc[ind])  # Display the corresponding DataFrame record
    print("")
```
